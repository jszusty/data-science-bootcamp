```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generowanie danych x od 0 do 2π z krokiem 0.1
x = np.arange(0, 2*np.pi, 0.1)

# Obliczenie wartości funkcji sin(x)
y = np.sin(x)

# Używanie stylu Seaborn dla wykresu
sns.set()

# Tworzenie wykresu
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Wykres funkcji y = sin(x)')
plt.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# Dane z ostatnich 10 notowań akcji Amazona
data = {
    'Data': ['2023-03-01', '2023-03-02', '2023-03-03', '2023-03-04', '2023-03-05',
             '2023-03-06', '2023-03-07', '2023-03-08', '2023-03-09', '2023-03-10'],
    'Cena': [3500, 3550, 3525, 3580, 3600, 3620, 3595, 3610, 3645, 3660]
}

# Tworzenie ramki danych pandas
df = pd.DataFrame(data)

# Tworzenie tabeli Plotly
table = go.Table(
    header=dict(values=['Data', 'Cena']),
    cells=dict(values=[df['Data'], df['Cena']])
)

# Tworzenie układu i wyświetlanie tabeli
layout = go.Layout(
    title='Ostatnie 10 notowań akcji Amazona',
)

fig = go.Figure(data=table, layout=layout)
fig.show()
```
